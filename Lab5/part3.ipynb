```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

z = x^0,25 + y^
0,25

```python
fig = plt.figure(figsize=(20, 16))

Z1 = np.power(X, 0.25) + np.power(Y, 0.25)

ax1 = fig.add_subplot(2, 3, 1, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
```

z = x^2-y^2

```python
Z2 = X ** 2 - Y ** 2

fig = plt.figure(figsize=(20, 16))

ax2 = fig.add_subplot(2, 3, 1, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='viridis')
```

 z = 2 * x + 3 * y

```python
Z3 = 2 * X + 3 * Y

fig = plt.figure(figsize=(20, 16))

ax3 = fig.add_subplot(2, 3, 1, projection='3d')
ax3.plot_surface(X, Y, Z3, cmap='viridis')
```

z = 2 + 2 * x + 2 * y - x^2 - y^2

```python
Z4 = 2 + 2 * X + 2 * Y - np.power(X, 2) - np.power(Y, 2)

fig = plt.figure(figsize=(20, 16))

ax4 = fig.add_subplot(2, 3, 1, projection='3d')
ax4.plot_surface(X, Y, Z4, cmap='viridis')
```
