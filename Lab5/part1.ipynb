Массив 10x10 со случайными значениями, его минимум и максимум

```python
random_array = np.random.randint(0, 100, (10, 10))
print(random_array)
max = random_array.max()
min = random_array.min()
print(f'max: {max}')
print(f'min: {min}')
```

Случайный вектор размера 30 и его среднее значение всех элементов

```python
vector = np.random.randint(0, 10, 30)
print(vector)
average = vector.mean()
f"Среднее значение: {average}"
```

8x8 матрицу и заполнить ее в шахматном порядке

```python
matrix = np.zeros((8, 8))
matrix[1::2, ::2] = 1
matrix[::2, 1::2] = 1
matrix
```

Заменить максимальный элемент на ноль


```python
array = np.array([1, 5, 3, 8, 2, 7])
array[array.argmax()] = 0
array
```

Найти наиболее частое значение в массиве

```python
array = np.array([1, 1, 1, 2, 2, 2, 2, 3, 3])
unique, counts = np.unique(array, return_counts=True)
unique[counts.argmax()]
```

Найти n наибольших значений в массиве

```python
n = 4
array = np.array([5, 1, 8, 3, 9, 2, 7, 4, 6])
largest_nums = np.partition(array, -n)[-n:]
largest_nums
```
